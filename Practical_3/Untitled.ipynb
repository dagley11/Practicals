```python
import time
jsons = []

for i in range(len(artists)):
    print(i)
    if i < 0:
        continue
    web = "http://musicbrainz.org/ws/2/artist/?query=arid:{}&fmt=json".format(artists[i])
    r = requests.get(web)
    try:
        while True:
            r = requests.get(web)
            json = r.json()
            error = list(json.keys())[0]
            if error != 'error':
                jsons.append(json)
                break
            time.sleep(1.1)
    except:
        jsons.append('NULL')
        continue
```
